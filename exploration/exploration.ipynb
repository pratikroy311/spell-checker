# read the data

```python
#this will contain all the words from text
words = []
with open("../data/shakespear_text.text",'r') as f:
    file = f.read()
    file = file.lower()
    words = re.findall(r'\w+',file)
print(words[:10])
```

```python
#let's create a vocabulary
vocab=set(words)
word_freq = {}
word_freq= Counter(words)
print("Top 10 most common words",word_freq.most_common()[:10])
```

```python
#relative frequency words
probs = {}
total = sum(word_freq.values())
for k in word_freq.keys():
    probs[k]= word_freq[k]/total

print(probs)
```

# finding similar words

```python
def auto_correct(word):
    word = word.lower()
    if word in vocab:
        return "Your word is correct"
    else:
        similarity = [1-(textdistance.Jaccard(qval=2).distance(v,word)) for v in word_freq.keys()]
        df = pd.DataFrame.from_dict(probs,orient='index').reset_index()
        df = df.rename(columns={'index':'Word',0:'Probability'})
        df['Similarity']= similarity
        output = df.sort_values(['Similarity','Probability'],ascending=False).head()
        return(output)
```

```python
#let's test this
auto_correct("preroative")
```
